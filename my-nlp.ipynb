```python
def stem(word):
    suffixes=["ing","ed","ly","s"]
    for suf in suffixes:
        if word.endswith(suf):
            return word[:-len(suf)]
    return word
```

```python
text = "playing played happily runs"
stemmed=[stem(word) for word in text.split()]
print(stemmed)
```

Rule_Based_lemmatization

```python
irregular = {
    "ate": "eat",
    "went": "go",
    "mice": "mouse",
    "better": "good"
}
def lemmatize(word):
    word=word.lower()
    if word in irregular:
        return irregular[word]
    if word.endswith("s"):
        return word[:-1]
    return word
```

```python
text = "cats ate cheese"
lemmas=[lemmatize(w) for w in text.split()]
print(lemmas)
```

TF-IDF (From Scratch)

```python
docs = [
    "machine learning is fun",
    "machine learning is powerful",
    "deep learning is a branch of machine learning"
]
```

```python
vocab = sorted(set(word for doc in docs for word in doc.split()))
print("Vocabulary:", vocab)
```

```python
def tf(word, doc):
    words = doc.split()
    return words.count(word) / len(words)
```

```python
import math

def idf(word, docs):
    N = len(docs)
    df = sum(1 for doc in docs if word in doc.split())
    return math.log(N / (df))
```

```python
tfidf_matrix = []

for doc in docs:
    tfidf_vector = [tf(word, doc) * idf(word, docs) for word in vocab]
    tfidf_matrix.append(tfidf_vector)

print("TF-IDF Matrix:")
for row in tfidf_matrix:
    print(row)
```

```python
def pos_tag(word):
    word = word.lower()

    # Pronouns
    pronouns = ["i", "you", "he", "she", "it", "we", "they"]
    
    # Determiners
    determiners = ["the", "a", "an", "this", "that", "these", "those"]
    
    # Prepositions
    prepositions = ["in", "on", "at", "by", "with", "to", "from", "into"]
    
    # Conjunctions
    conjunctions = ["and", "or", "but", "because", "so", "yet"]

    # Numbers
    if word.isdigit():
        return "NUM"
    
    # Pronouns
    if word in pronouns:
        return "PRON"
    
    # Determiners
    if word in determiners:
        return "DET"
    
    # Prepositions
    if word in prepositions:
        return "PREP"
    
    # Conjunctions
    if word in conjunctions:
        return "CONJ"
    
    # Adverbs (ending in -ly)
    if word.endswith("ly"):
        return "ADV"
    
    # Verbs (ending in -ing or -ed)
    if word.endswith("ing") or word.endswith("ed"):
        return "VERB"
    
    # Adjectives (ending in -ous, -ful, -ive)
    if word.endswith("ous") or word.endswith("ful") or word.endswith("ive"):
        return "ADJ"
    
    # Default: NOUN
    return "NOUN"


# ------------------------------
# Test the POS tagger
# ------------------------------

sentence = "The quick brown fox is running quickly into the forest"
words = sentence.split()

tags = [(w, pos_tag(w)) for w in words]

print(tags)
```

```python
training_data = [
    ("Win money now", "spam"),
    ("Claim your free prize", "spam"),
    ("Limited offer just for you", "spam"),
    ("Are we still meeting today?", "ham"),
    ("Let's have lunch tomorrow", "ham"),
    ("Can you send me the report?", "ham")
]
```

```python
def tokenize(text):
    return text.lower().split()
```

```python
word_counts = {"spam": {}, "ham": {}}
class_counts = {"spam": 0, "ham": 0}

for text, label in training_data:
    class_counts[label] += 1
    for word in tokenize(text):
        word_counts[label][word] = word_counts[label].get(word, 0) + 1
```

Step 4: Prediction Function

Formula:

ùëÉ ( ùëê ùëô ùëé ùë† ùë† ‚à£ ùë° ùëí ùë• ùë° ) ‚àù ùëÉ ( ùëê ùëô ùëé ùë† ùë† ) √ó Œ† ùëÉ ( ùë§ ùëú ùëü ùëë ‚à£ ùëê ùëô ùëé ùë† ùë† ) P(class‚à£text)‚àùP(class)√óŒ†P(word‚à£class)

We use Laplace smoothing:

ùëÉ ( ùë§ ùëú ùëü ùëë ‚à£ ùëê ùëô ùëé ùë† ùë† )
count(word) + 1 total words in class + ùë£ ùëú ùëê ùëé ùëè ùë¢ ùëô ùëé ùëü ùë¶ ùë† ùëñ ùëß ùëí P(word‚à£class)= total words in class+vocabularysize count(word)+1‚Äã

```python
def predict(text):
    words = tokenize(text)
    scores = {"spam": 1, "ham": 1}
    
    vocab = set()
    for c in word_counts:
        vocab.update(word_counts[c].keys())
    vocab_size = len(vocab)

    for c in ["spam", "ham"]:
        total_words = sum(word_counts[c].values())
        prior = class_counts[c] / sum(class_counts.values())
        scores[c] *= prior

        for w in words:
            word_prob = (word_counts[c].get(w, 0) + 1) / (total_words + vocab_size)
            scores[c] *= word_prob

    return max(scores, key=scores.get)
```

```python
print(predict("free money offer for you"))
print(predict("are we meeting tomorrow?"))
print(predict("claim your prize now"))
print(predict("send me the report"))
```
