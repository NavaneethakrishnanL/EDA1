```python
def stem(word):
    suffixes=["ing","ed","ly","s"]
    for suf in suffixes:
        if word.endswith(suf):
            return word[:-len(suf)]
    return word
```

```python
text = "playing played happily runs"
stemmed=[stem(word) for word in text.split()]
print(stemmed)
```

Rule_Based_lemmatization

```python
irregular = {
    "ate": "eat",
    "went": "go",
    "mice": "mouse",
    "better": "good"
}
def lemmatize(word):
    word=word.lower()
    if word in irregular:
        return irregular[word]
    if word.endswith("s"):
        return word[:-1]
    return word
```

```python
text = "cats ate cheese"
lemmas=[lemmatize(w) for w in text.split()]
print(lemmas)
```

Bag of Words (BoW) â€“ From Scratch

```python
docs = [
    "NLP is fun",
    "NLP is powerful",
    "NLP is the future"
]

vocab=sorted(set(word for doc in docs for word in doc.split()))
print("Vocabulary: ",vocab)
```

```python
vectors=[]
for doc in docs:
    words=doc.split()
    vector=[words.count(word) for word in vocab]
    vectors.append(vector)
print("BOW_Vectors: ")
for v in vectors:
    print(v)
```

TF-IDF (From Scratch)

```python
docs = [
    "machine learning is fun",
    "machine learning is powerful",
    "deep learning is a branch of machine learning"
]
```

```python
vocab = sorted(set(word for doc in docs for word in doc.split()))
print("Vocabulary:", vocab)
```

```python
def tf(word, doc):
    words = doc.split()
    return words.count(word) / len(words)
```

```python
import math

def idf(word, docs):
    N = len(docs)
    df = sum(1 for doc in docs if word in doc.split())
    return math.log(N / (df))
```

```python
tfidf_matrix = []

for doc in docs:
    tfidf_vector = [tf(word, doc) * idf(word, docs) for word in vocab]
    tfidf_matrix.append(tfidf_vector)

print("TF-IDF Matrix:")
for row in tfidf_matrix:
    print(row)
```
